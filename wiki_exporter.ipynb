```python
import pandas as pd
```

```python
url = "https://en.wikipedia.org/wiki/Biblical_literalist_chronology"
```

```python
# Import tables from the URL
tables = pd.read_html(url)
```

```python
table_titles = {
    8: "Adam to the Flood",
    9: "The Flood to Abraham",
    10: "Abraham to Joseph",
    11: "Egypt to the Exodus",
    12: "The Wilderness Period to the Conquest of Canaan",
    13: "The Judges to the United Monarchy",
    14: "The Divided Monarchy to the Destruction of the Temple",
    15: "The Babylonian Captivity to the Decree of Cyrus",
    16: "The Second Temple to Alexander the Great",
    17: "Jaddua the high priest to John Hyrcanus",
    18: "Esther 11:1 – the 4th year of Ptolemy and Cleopatra",
    19: "2 Maccabees 1:10–12 – Aristobulus II"
}
```

```python
tables[17]
```

```python
# Enumerate table[8] line by line:
for table_id, title in table_titles.items():
    print(f"| Date | Literal biblical chronology, events {title} | Bible reference |")
    print("| --- | ------- | ----------- |")
    for i, row in enumerate(tables[table_id].iterrows()):
        # Check if iloc[0] is NaN
        if row[1].iloc[0] is None:
            continue
        try:
            date = row[1].iloc[0]
            date = date.replace("c. ", "").split(" ")
        except AttributeError:
            continue
        if date[0] == "c.":
            date = date[1]
        else:
            date = date[0]
        if date == "no":
            continue
        date = date.split("[")[0]
        # di = date.split('/')[0]
        date = date.replace("-", " BC - ")
        date = date.replace("–", " BC - ")
        print(f"| {date} BC | {row[1].iloc[1]} | {row[1].iloc[2]} |")
    print()
```

```python
url = "https://en.wikipedia.org/wiki/Timeline_of_Chinese_history"
tables = pd.read_html(url)
```

```python
valid_months = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]
print(f"| Date | Chinese history event |")
print("| --- | ------- |")
for index in range(11,55):
    for i, row in enumerate(tables[index].iterrows()):
        # Check if iloc[0] is NaN
        year = row[1].iloc[0]
        if year is None:
            continue
        bc=False
        year_end = None
        if isinstance(year, int) is False:
            if year.endswith("BC"):
                year = year.replace("BC", "").replace(",","").strip()
                year = int(year)
                bc = True
            else:
                if '-' in year:
                    years = year.split("-") 
                    year = int(years[0].split(" ")[0])
                    year_end = int(years[1].split(" ")[0])
                else:
                    year = year.split(" ")[0]
                    year = int(year)
        if index>11 and row[1].iloc[1] is not None and isinstance(row[1].iloc[1], str):
            # print(row[1].iloc[1])
            month_day = row[1].iloc[1].split(" ")
            if len(month_day) == 1:
                month = month_day[0]
                day = None
            else:
                month = month_day[1]
                day = month_day[0]
            if month not in valid_months:
                month = None
                day = None
            else:
                # get index of month:
                month = valid_months.index(month) + 1
                # check if day is a number
                if day is None or not day.isdigit():
                    day = None
                    date = f"{year}-{month:02d}"
                else:
                    day = int(day)
                    date = f"{year}-{month:02d}-{day:02d}"
        else:
            month = None
            day = None
            if year_end is not None:
                date = f"{year} - {year_end}"
            else:
                date = f"{year}"
        if bc:
            date = f"{date} BC"
        if index == 11:
            print(f"| {date} | {row[1].iloc[1]} |")
        else:
            if row[1].iloc[2] is not None:
                print(f"| {date} | {row[1].iloc[2]} |")
```

```python
url = "https://en.wikipedia.org/wiki/List_of_large_volcanic_eruptions"
tables = pd.read_html(url)
```

```python
entries = []
for index in range (1, 7):
    for i, row in enumerate(tables[index].iterrows()):
        if isinstance(row[1].iloc[4], str):
            if row[1].iloc[4].startswith("^"):
                continue
            date_r = row[1].iloc[4].replace("+", "")
            date = date_r.split("[")[0]
            date = date.replace("–", "-")
            dates = date.split("-")
        else:
            date = row[1].iloc[4]
            dates = [date]
        if len(dates) == 1:
            #print(row)
            #print(index, i, dates)
            date = float(dates[0])
            end_date = None
        else:
            date = float(dates[0])
            end_date = float(dates[1])
        if index > 4:
            date = date * 1000  # Ma
            if end_date is not None:
                end_date = end_date * 1000
        if end_date is not None:
            date = f"{date} kya BP - {end_date} kya BP"
        else:
            date = f"{date} kya BP"
        desc = f"{row[1].iloc[1]}, {row[1].iloc[2]}, {row[1].iloc[5]}"
        vei = row[1].iloc[0].replace('?','')
        volume = row[1].iloc[3].replace('?','')
        entries.append(f"| {date} | {desc} | {vei} | {volume} |")

print(f"| Date | Volcanic eruption | VEI | Material Volume (km3) |")
print("| --- | ------- | --- | --------------------- |")
entries.reverse()
for entry in entries:
    print(entry)
```

```python
url = "https://en.wikipedia.org/wiki/List_of_large_Holocene_volcanic_eruptions"
tables = pd.read_html(url)
```

```python
tables[0]
```

```python
valid_months = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]

print(f"| Date | Volcanic eruption | VEI | Material Volume (km3) |")
print("| --- | ------- | --- | --------------------- |")
for index in range (0, 3):
    entries = [[], [], []]
    for i, row in enumerate(tables[index].iterrows()):
        vei = row[1].iloc[0]
        desc = row[1].iloc[1] + ", " + row[1].iloc[2]
        if index != 1:
            add_desc = row[1].iloc[6]
        else:
            add_desc = row[1].iloc[7]
        if add_desc != None and isinstance(add_desc, str):
            desc = desc + ", " + add_desc
        volume = row[1].iloc[3]
        if index < 2:
            raw_date = row[1].iloc[5]
            bc = False
            if ' BCE' in raw_date:
                raw_date = raw_date.replace(" BCE", "")
                bc = True
            raw_date = raw_date.replace("c.", "").replace("–", "-").replace("?", "").replace(" CE", "")
            raw_date = raw_date.split("[")[0]
            if len(raw_date.split("-")) == 3:
                date = raw_date
            else:
                if "," in raw_date:
                    dates = raw_date.split(",")
                    month_sym, day = dates[0].split(" ")
                    year = dates[1].strip()
                    month = valid_months.index(month_sym) + 1
                    if '-' in day:
                        days = day.split("-")
                        day = int(days[0])
                        day_end = int(days[1])
                    else:
                        day = int(day)
                        day_end = None
                    if day_end is not None:
                        date = f"{year}-{month:02d}-{day:02d} - {year}-{month:02d}-{day_end:02d}"
                    else:
                        date = f"{year}-{month:02d}-{day:02d}"
                else:
                    if '±' in raw_date:
                        date_parts = raw_date.split("±")
                        delta = int(date_parts[1])
                        date_med = int(date_parts[0])
                        if bc:
                            date = f"{date_med-delta} BC - {date_med+delta} BC"
                        else:
                            date = f"{date_med-delta} - {date_med+delta}"
                    else:
                        if bc:
                            date = f"{raw_date} BC"
                        else:
                            date = raw_date
        else:
            continue
        entries[index].append(f"| {date} | {desc} | {vei} | {volume} |")
    for index in range(3):
        entries[index].reverse()
        for entry in entries[index]:
            print(entry)
        print("-----")
```
